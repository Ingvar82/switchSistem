```python
from nornir import InitNornir
from nornir.core.task import Task, Result
from nornir_utils.plugins.functions import print_result,print_title
from nornir_netmiko import netmiko_send_command, netmiko_send_config

path = ('C:/Users/KOMP/Desktop/DOC/MAIN/Nornir/')
nr = InitNornir(config_file=path+"config.yaml")

def lb_create(lb_task):
    lb_task.run(netmiko_send_config,config_commands=["no ip route 0.0.0.0 0.0.0.0"])
    lb_task.run(netmiko_send_command, command_string="show cur")
    
result = nr.run(lb_create)
    
print_title("NETMIKO DEMO")
print_result(result)
```
